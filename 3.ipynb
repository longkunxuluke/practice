```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('housing.data',header=None,sep= '\s+')
df.columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
pd.DataFrame(df.head())
print(pd.DataFrame(df.head()))
```

```python
print('df-shape', df.shape)
x = df[:20][['RM']].values
y = df[:20]['MEDV'].values
print('x3:\n',x[:3])
print('y3:\n',y[:3])
model = LinearRegression()
model.fit(x,y)
print(model)
```

```python
print('model.coef:%.2f'% model.coef_)
print('model.intercept:%.2f'% model.intercept_)
```

```python
new_data = np.array([[6]])
print(model.predict(new_data))
```

```python
x_plt = np.arange(5,9,1)
print('1Darray',x_plt)
```

```python
x_plt=np.arange(5,9,1)[:,np.newaxis]
print('2Darray',x_plt)
```

```python
y_pred = model.predict(x_plt)
print(y_pred)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x,y,color='blue',label='data')
plt.plot(x_plt,y_pred,color='black',linestyle='-',label='LinearRegression')
plt.ylabel('Price in $1000s [MEDV]')
plt.xlabel('Average number of rooms [RM]')
plt.title('Boston house-prices')
plt.legend(loc='lower right')
plt.show()
```
